```python
contacts = {}

def add_contact():
    name = input("Name: ")
    phone = input("Phone: ")
    contacts[name] = phone
    print("Contact added successfully!")

def search_contact():
    name = input("Name: ")
    print(contacts.get(name, "Not found"))

def delete_contact():
    name = input("Name: ")
    if name in contacts:
        contacts.pop(name)
        print("Contact deleted!")
    else:
        print("Contact not found!")

while True:
    print("\n1. Add  2. Search  3. Delete  4. Exit")
    ch = input("Choice: ")

    if ch == "1":
        add_contact()
    elif ch == "2":
        search_contact()
    elif ch == "3":
        delete_contact()
    elif ch == "4":
        print("Exiting program...")
        break
    else:
        print("Invalid choice!")
```

```python
tasks = []

def add_task():
    tasks.append(input("Task: "))
    print("Task added!")

def view_tasks():
    if not tasks:
        print("No tasks available!")
    else:
        for i, t in enumerate(tasks, 1):
            print(i, t)

def remove_task():
    view_tasks()
    if not tasks:
        return
    idx = int(input("Task no: "))
    if 1 <= idx <= len(tasks):
        tasks.pop(idx - 1)
        print("Task removed!")
    else:
        print("Invalid task number!")

while True:
    print("\n1. Add  2. View  3. Remove  4. Exit")
    ch = input("Choice: ")

    if ch == "1":
        add_task()
    elif ch == "2":
        view_tasks()
    elif ch == "3":
        remove_task()
    elif ch == "4":
        print("Goodbye 👋")
        break
    else:
        print("Invalid choice!")
```

```python
balance = 10000

def check_balance():
    print("Balance:", balance)

def deposit():
    global balance
    amount = int(input("Amount: "))
    balance += amount
    print("Deposited successfully!")

def withdraw():
    global balance
    amt = int(input("Amount: "))
    if amt <= balance:
        balance -= amt
        print("Withdrawal successful!")
    else:
        print("Insufficient balance!")

while True:
    print("\n1. Balance  2. Deposit  3. Withdraw  4. Exit")
    ch = input("Choice: ")

    if ch == "1":
        check_balance()
    elif ch == "2":
        deposit()
    elif ch == "3":
        withdraw()
    elif ch == "4":
        print("Thank you for using ATM 👋")
        break
    else:
        print("Invalid choice!")
```

```python
questions = [
    {"q": "Capital of India?", "a": "delhi"},
    {"q": "2+3?", "a": "5"}
]

score = 0

for q in questions:
    if input(q["q"] + " ").lower() == q["a"].lower():
        score += 1

print("Score:", score)
```

```python
expenses = {}

def add_expense():
    cat = input("Category: ")
    amt = int(input("Amount: "))
    expenses[cat] = expenses.get(cat, 0) + amt
    print("Expense added!")

def view_expenses():
    if not expenses:
        print("No expenses recorded!")
    else:
        for c, a in expenses.items():
            print(c, ":", a)

while True:
    print("\n1. Add  2. View  3. Exit")
    ch = input("Choice: ")

    if ch == "1":
        add_expense()
    elif ch == "2":
        view_expenses()
    elif ch == "3":
        print("Exiting Expense Tracker 👋")
        break
    else:
        print("Invalid choice!")
```
