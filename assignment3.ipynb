```python
scores = [60, 70, 80, 90]
total = sum(scores)
average = total / len(scores)
count = 0
if scores[0]>average:
   count+=1
if scores[1]>average:
   count+=1
if scores[2]>average:
   count+=1
if scores[3]>average:
   count+=1

print("total score:",total)
print("average score:",average)
print("students above average:",count)
```
