```python
import numpy as np
import pandas as pd

data = {
    "Name": ["Ravi","Rahul","akshata","sneha","Anu","John","bob","Priya"],
    "Age": [25,30,22,np.nan,29,25,45,120],
    "Salary": [50000,np.nan,45000,60000,52000,50000,150000,30000],
    "Department": ["IT","HR","IT","Finance","IT","IT","Management","HR"],
    "Experience": [2,5,1,4,3,2,20,1]
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df)

```

```python
df.fillna({
    "Age": df["Age"].mean(),
    "Salary": df["Salary"].mean()
}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df=df[df["Age"]<100]
print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Department"] = le.fit_transform(df["Department"])

df.head()

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["Age", "Salary", "Experience"]] = scaler.fit_transform(
    df[["Age", "Salary", "Experience"]]
)
print(df)
```

```python
print("\n Cleaned Data:\n")
print(df)
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Salary"],bins=5)
plt.title("Salary Distribution after Cleaning")
plt.show()
```
